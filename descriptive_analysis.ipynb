```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = 'data/1d_data/'
df = pd.read_parquet(path+'BTCUSDT.parquet')
df['close'] = pd.to_numeric(df['close'])
df['open'] = pd.to_numeric(df['open'])
```

```python
df['return'] = (df['close'] - df['open'])/df['open']
df['return'] = np.log(df['close']) - np.log(df['open'])
df['return'] = ((df['close'] - df['open'])*100) / df['open']
df['normalized_return'] = (df['return'] - np.mean(df['return'])) / np.sqrt( np.mean(df['return']*df['return']) - (np.mean(df['return']))**2)
```

```python
sns.set()
```

```python
plt.figure(figsize=(12,6))
plt.title('Log Returns')
plt.plot(df['normalized_return'], '-');
```
